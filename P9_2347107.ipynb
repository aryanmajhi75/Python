```python
acousticness=data['acousticness'].mean()
duration_ms=data['duration_ms'].mean()
energy=data['energy'].mean()
instrumentalness=data['instrumentalness'].mean()
liveness=data['liveness'].mean()
speechiness=data['speechiness'].mean()
valence=data['valence'].mean()
```

```python
data['acousticness'].replace(np.nan,acousticness, inplace=True)
data['duration_ms'].replace(np.nan,duration_ms, inplace=True)
data['energy'].replace(np.nan,energy, inplace=True)
data['instrumentalness'].replace(np.nan,instrumentalness, inplace=True)
data['liveness'].replace(np.nan,liveness, inplace=True)
data['speechiness'].replace(np.nan,speechiness, inplace=True)
data['valence'].replace(np.nan,valence, inplace=True)

data.isna().sum()
```

Null values removed using mean()

```python
numeric_columns = data.select_dtypes(include=['number'])
correlation = numeric_columns.corr()
sns.heatmap(correlation, annot=True)
print(correlation)
```

```python
threshold = 0.7

high_correlation_columns = set()
for i in range(len(correlation.columns)):
    for j in range(i):
        if abs(correlation.iloc[i, j]) > threshold:
            col1 = correlation.columns[i]
            col2 = correlation.columns[j]
            high_correlation_columns.add(col1)
            high_correlation_columns.add(col2)

# Print the columns with high correlation
print("Columns with high correlation:")
print(high_correlation_columns)
```

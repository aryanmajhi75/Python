Q1 Given the list of CGPA of students CGPA=[9,7,6,5,10,3,9,6,6,4]. Write

a) An expression that evaluates to the number of 6 CGPA

b) A statement that changes the last CGPA to 3

c) An expression that evaluates the maximum CGPA

d) A statement that sorts the list CGPA

e) An expression that evaluates the average CGPA

```python
cgpa=[9,7,6,5,10,3,9,6,6,4]
```

```python
num6=[x for x in cgpa if x==6]
print(num6)
```

```python
cgpa[len(cgpa)-1]=3
cgpa
```

```python
max=cgpa[0]
max=[x for x in cgpa if x>max]
print(max)
```

```python
cgpa.sort()
print(cgpa)
```

```python
def average():
  sum=0
  avg=0.0
  [sum:=sum+x for x in cgpa]
  avg=sum/len(cgpa)
  print(avg)

average()
```
